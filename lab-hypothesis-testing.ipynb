```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
# We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. 
# Choose the propper test and, with 5% significance, comment your findings.

# We will use a two-sample t-test to compare the means of the two groups. 

# H0: The null hypothesis is that the means are equal
# H1: while the alternative hypothesis is that the mean of the Dragon group is greater than the mean of the Grass group.

# Muestras: We will first filter the data to get the two groups we are interested in.

df_dragon = df[df["Type 1"] == "Dragon"]
df_grass = df[df["Type 1"] == "Grass"]

# We will now perform the t-test.

t_stat, p_val = st.ttest_ind(df_dragon["HP"], df_grass["HP"], equal_var = False)

# We will now check if the p-value is less than 0.05.

if p_val < 0.05:
    print("We reject the null hypothesis.")
else:
    print("We fail to reject the null hypothesis.")

# We reject the null hypothesis. This means that we have enough evidence to say that the mean of the Dragon group is greater than the mean of the Grass group.

# We can also calculate the mean of the two groups to confirm this.

mean_dragon = df_dragon["HP"].mean()
mean_grass = df_grass["HP"].mean()

print("Mean of Dragon group:", mean_dragon)
print("Mean of Grass group:", mean_grass)
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind

# Cargar el dataset
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv"
df = pd.read_csv(url)

# Coordenadas de la escuela y el hospital
school_coords = (-118, 34)
hospital_coords = (-122, 37)

def euclidean_distance(lat, lon, ref_lat, ref_lon):
    return np.sqrt((lat - ref_lat) ** 2 + (lon - ref_lon) ** 2)
```

```python
# Calcular distancia a la escuela y al hospital (distancia directa entre dos puntos)
df['dist_school'] = df.apply(lambda row: euclidean_distance(row['latitude'], row['longitude'], school_coords[1], school_coords[0]), axis=1)
df['dist_hospital'] = df.apply(lambda row: euclidean_distance(row['latitude'], row['longitude'], hospital_coords[1], hospital_coords[0]), axis=1)

# Definir si la casa está cerca de una escuela u hospital
df['close_to_facility'] = (df['dist_school'] < 0.50) | (df['dist_hospital'] < 0.50)

# Separar las casas en cercanas y lejanas
df_close = df[df['close_to_facility']]
df_far = df[~df['close_to_facility']]

# Planteamos H0: Las casas cercanas a escuelas u hospitales no son más caras
# Planteamos H1: Las casas cercanas a escuelas u hospitales son más caras

# Selección del tipo de test:
# - Seleccionamos un test t de dos muestras
# - No asumimos varianzas iguales
# - Nivel de significancia 0.05

# Prueba de hipótesis (t-test)
t_stat, p_value = ttest_ind(df_close['median_house_value'], df_far['median_house_value'], equal_var=False)

# Resultados
print("T-Statistic:", t_stat)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Conclusión: Hay evidencia estadística para afirmar que las casas cercanas a escuelas u hospitales son más caras.")
else:
    print("Conclusión: No hay evidencia suficiente para afirmar que la cercanía a escuelas u hospitales afecta el precio.")
```
